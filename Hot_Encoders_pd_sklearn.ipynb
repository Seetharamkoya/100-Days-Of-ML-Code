## One Hot encoading where we are assinning a single colum for each 

```python
dummies = pd.get_dummies(df.town)
dummies
```

## Concat method,  Merge the 2 data frames either colums or rows 

```python
merged = pd. concat([df, dummies], axis = 'columns')
merged
```

```python
final = merged .drop(['town', 'west windsor'], axis = 'columns')
final
```

## Importing Machine learning model and Object

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
y
```

## Traing the data

```python
model.fit(X,y)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
